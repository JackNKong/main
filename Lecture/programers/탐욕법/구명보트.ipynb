```python
def solution(people: list, limit: int) -> int:
    i = 0
    j = len(people) - 1
    boat = 0
    people.sort()
    while not(i > j):
        if (people[i] + people[j] > limit):
            boat = boat + 1
            j = j - 1
        else:
            if not(len(people) == 1): 
                boat = boat + 1
                j = j - 1
                i = i + 1
            else:
                boat = boat + 1
    return boat


```
