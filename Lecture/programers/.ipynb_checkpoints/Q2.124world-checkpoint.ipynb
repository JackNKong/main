```python
def next_num(n):
    if n =="1" : return "2"
    elif n =="2" : return "4"
    else : return "1"



def solution(n):
    answer = ''
    answer_list=['1']
    for i in range(1,n):
        if set(answer_list) =={"4"} :
            answer_list.append("1")
            answer_list=["1" for i in answer_list]
        elif answer_list[-1] == "4" and len(answer_list) >=2:
            i_list = []
            for i,txt in enumerate(answer_list):
                #print(i,txt)
                if txt =="4" : 
                    answer_list[i]="1"
                    i_list.append(i)
                
            min_i =min(i_list)
            answer_list[min_i-1] = next_num(answer_list[min_i-1])
            #answer_list[-1] = next_num(answer_list[-1])
            #answer_list[-2] = next_num(answer_list[-2])
        else : 
            answer_list[-1] = next_num(answer_list[-1])
        
    answer = ''.join(answer_list)
        
        
    return answer
```

```python
for i in range(0,100) :
    print(solution(i))
#solution(17)
```
