```python
from googleapiclient.discovery import build
api_key= '#####'
youtube= build('youtube','v3',developerKey=api_key)

#https://github.com/googleapis/google-api-python-client
```

```python
#코코몽 id 

requests = youtube.channels().list(
    part='id',
    forUsername='1Cocomong',
    maxResults=5
)
response= requests.execute()
channel_id = response['items'][0]['id']
channel_id
```

```python
requests = youtube.playlists().list(
    channelId=channel_id,
    part='snippet,id',
    maxResults=50
)
response= requests.execute()
```

```python
playlist_list= response['items']
playlist_list[0]
```

```python
datas=[]
for playlist in playlist_list[:1] : 
    playlist_id = playlist['id']
    playlist_nm = playlist['snippet']['title']
    requests = youtube.playlistItems().list(
    playlistId=playlist_id,
    part='contentDetails',
    maxResults=50
    )
    response= requests.execute()
    video_list = response['items']
    for video in video_list[:1] : 
        video_id =video['contentDetails']['videoId']
        requests = youtube.videos().list(
            id=video_id,
            part='snippet,statistics,contentDetails',
            maxResults=50
        )
        response= requests.execute()
        if len(response['items']) != 0 :
            tmp_dic={}
            tmp_dic['playlist_id'] = playlist_id
            tmp_dic['playlist_nm'] = playlist_nm
            tmp_dic['video_id'] = video_id
            tmp_dic['video_nm'] = response['items'][0]['snippet']['title']
            if 'commentCount' in response['items'][0]['statistics'].keys() : 
                tmp_dic['comment_cnt'] = response['items'][0]['statistics']['commentCount']
            if 'dislikeCount' in response['items'][0]['statistics'].keys() : 
                tmp_dic['dislike_cnt'] = response['items'][0]['statistics']['dislikeCount']

            if 'favoriteCount' in response['items'][0]['statistics'].keys() : 
                tmp_dic['favorite_cnt'] = response['items'][0]['statistics']['favoriteCount']
            if 'likeCount' in response['items'][0]['statistics'].keys() : 
                tmp_dic['like_cnt'] = response['items'][0]['statistics']['likeCount']
            if 'viewCount' in response['items'][0]['statistics'].keys() :
                tmp_dic['view_cnt'] = response['items'][0]['statistics']['viewCount']
            datas.append(tmp_dic)

        
```
